```python
import matplotlib.pyplot as plt
```

```python
john_results = [83.05, 91.43]
justin_results = [87.85, 92.45]
angel_results = [82.21, 91.24]
ensamble_results = [96.12]
final_results = john_results + justin_results + angel_results + ensamble_results
labels = ['John', 'John w/ Optuna', 'Justin', 'Justin w/ Optuna', 'Angel', 'Angel w/ Optuna', 'Ensemble']

for i in range(len(final_results)):
    text = '-'
    for j in range(101 - int(final_results[i])):
        text += '-' 
    text += str(final_results[i]) + '%'
    plt.text(final_results[i], i, text)

plt.title("Model Performance")
plt.xlabel("Val Accuracy %")
plt.ylabel("Models")
plt.barh(labels, final_results)
```
